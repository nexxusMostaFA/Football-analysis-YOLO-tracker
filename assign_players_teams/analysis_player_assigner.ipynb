```python
from sklearn.cluster import KMeans
import cv2
import matplotlib.pyplot as plt
```

```python
img_path = r'C:\Users\mostafa\Documents\GitHub\Football-analysis-YOLO-tracker\data\cropped_image.jpg'

image = cv2.imread(img_path)
plt.imshow(image)
```

```python
top_half_img = image[0 : int(image.shape[0]/2 ) , :]
plt.imshow(top_half_img)
```

```python
kmeans = KMeans(n_clusters=2 , random_state= 0 )

image_2d = top_half_img.reshape(-1 ,3)

kmeans.fit(image_2d)
```

```python
labels = kmeans.labels_
clustered_image = labels.reshape(top_half_img.shape[0] , top_half_img.shape[1])
plt.imshow(clustered_image)
```

```python
corner_points = [clustered_image[0,0] ,clustered_image[0,-1] ,clustered_image[-1,0] ,clustered_image[-1,-1]]

background = max(set(corner_points) , key=corner_points.count)
print("background = " , background)
player = 1- background
print("player = " , player)
```

```python
print("player color = " , kmeans.cluster_centers_[player])
print("background color = " , kmeans.cluster_centers_[background])
```
